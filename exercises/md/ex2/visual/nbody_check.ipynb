# To check if the code is correct

```python
import numpy as np  
import matplotlib.pyplot as plt

import utils


xyz = [(0, 1), (0, 2), (1, 2)]
```

## Stable Circular Orbit 

### 2 particles

```python
filename = 'test_2part_stable_orbit'
pos, ids, mass, time = utils.read_data(filename)
```

```python
colors = utils.generate_colors(len(ids), cmap='gnuplot_r')
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
for i, particle in enumerate(ids):
    dist_in_fin = np.linalg.norm(pos[particle, :][0] - pos[particle, :][-1])
    for j in range(3):
        if dist_in_fin < 1e-3:
            ax[j].plot(pos[particle, xyz[j][0]], pos[particle, xyz[j][1]], 'o', color=colors[i], ms=20)
        else:
            ax[j].plot(pos[particle, xyz[j][0]], pos[particle, xyz[j][1]], '-', color=colors[i])
            ax[j].plot(pos[particle, xyz[j][0]][-1], pos[particle, xyz[j][1]][-1], 'o', color=colors[i], ms=10)

fig, ax = utils.labels_plots(fig, ax)
fig.savefig('../output/figures/' + filename + '.png')
```
